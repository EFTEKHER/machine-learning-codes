```python
import matplotlib.pyplot as plt
import seaborn as sns             
from sklearn.datasets import make_moons
import pandas as pd
```

```python
x,y=make_moons(n_samples=250,noise=0.05)
```

```python
df={"data1":x[:,0],"data2":x[:,1]}
```

```python
dataset=pd.DataFrame(df)
```

```python
sns.scatterplot(x="data1",y="data2",data=dataset)
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN(eps=0.3,min_samples=5)
```

```python
db.fit_predict(dataset)
```

```python
dataset['predict']=db.fit_predict(dataset)
```

```python
sns.scatterplot(x="data1",y="data2",data=dataset, hue="predict",palette="viridis")
```

```python
sns.scatterplot(x="data1",y="data2",data=dataset, hue=y,palette="viridis")
```
