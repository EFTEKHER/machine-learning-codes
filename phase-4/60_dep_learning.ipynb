```python
dataset=pd.read_csv('./Churn_Modelling.csv')
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,StandardScaler
```

```python
dataset.drop(columns=['RowNumber',	'CustomerId',	'Surname' ,'Geography',	'Gender'	],inplace=True)
```

```python
input_data=dataset.iloc[:,:-1]
output_data=dataset.iloc[:,-1]
```

```python
dataset
```

```python
ss=StandardScaler()
```

```python
input_data=pd.DataFrame(ss.fit_transform(input_data),columns=input_data.columns)
```

```python
from keras.layers import Dense
```

```python
from keras.models import Sequential
```

```python
input_data.shape
```

```python
ann=Sequential()
```

```python
ann.add(Dense(6,activation='relu',input_dim=8))
ann.add(Dense(4,activation='relu'))
ann.add(Dense(2,activation='relu'))
ann.add(Dense(1,activation='sigmoid'))

```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(input_data,output_data,test_size=0.2,random_state=21)
```

```python
ann.fit(x_train,y_train,batch_size=100,epochs=50)
```

```python
from sklearn.metrics import accuracy_score
```

```python
prd=ann.predict(x_test)
```

```python
prd_data=[]
for i in prd:
      if i>0.5:
        prd_data.append(1)
      else:
         prd_data.append(0)
        
      
    
```

```python
prd1=ann.predict(x_train)
prd_data_train=[]
for i in prd1:
      if i>0.5:
        prd_data_train.append(1)
      else:
         prd_data_train.append(0)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prd_data)
```

```python
accuracy_score(y_train,prd_data_train)
```
