```python
market=[]

for i in range(0, dataset.shape[0]):
    customer=[]
    for j in dataset.columns:
        if type(dataset[j][i])==str:
            customer.append(dataset[j][i])
    market.append(customer)        
    
```

```python
l=[]
for i in market:
    for j in i:
        l.append(j)
```
