```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv('placement-dataset.csv')
```

```python
dataset.head()
```

```python
dataset.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
sns.scatterplot(x='cgpa',y='iq',hue='placement',data=dataset)
```

```python
x=dataset.iloc[:,:-1]
```

```python
y=dataset['placement']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=11)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr=Perceptron(alpha=.02)
```

```python
pr.fit(x_train,y_train)
```

```python
pr.score(x_test,y_test)
```

```python
pr.score(x_train,y_train)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=pr)
```
