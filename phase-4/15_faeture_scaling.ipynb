```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
dataset=pd.read_csv("loan.csv")
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset["ApplicantIncome"].fillna(dataset["ApplicantIncome"].mean(),inplace=True)
```

```python
sns.displot(data=dataset["ApplicantIncome"],kde=True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
ss.fit(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"] =pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]),columns=["x"])
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
# Create a figure with 1 row and 2 columns of subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the first distribution
sns.distplot(dataset["ApplicantIncome"], ax=axes[0])
axes[0].set_title('Before')

# Plot the second distribution
sns.distplot(dataset["ApplicantIncome_ss"], ax=axes[1])
axes[1].set_title('After')

# Adjust the layout to prevent overlap
fig.tight_layout()

# Display the plots
plt.show()
```
