```python
import matplotlib.pyplot as plt
import numpy as np                   
import pandas as pd        
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
import seaborn as sns                        
```

```python
dataset=pd.read_csv('Placement.csv')
```

```python
dataset.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='cgpa',y='iq', data=dataset,hue='placement')

plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset['placement']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=21)
```

```python
fig1,ax1=plt.subplots(figsize=(6,6))
fig2,ax2=plt.subplots(figsize=(6,6))
fig3,ax3=plt.subplots(figsize=(6,6))

sns.kdeplot(data=dataset['cgpa'], label='cgpa',ax=ax1)
sns.kdeplot(data=dataset['iq'], label='iq',ax=ax2)

sns.kdeplot(data=dataset['placement'], label='placement',ax=ax3)

ax1.legend()
ax2.legend()
ax3.legend()

plt.show(fig1)
plt.show(fig2)
plt.show(fig3)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)
gnb.score(x_train,y_train)
```

```python
mnb=MultinomialNB()
```

```python
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)
```

```python
bnb=BernoulliNB()
```

```python
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)
```

```python
fig1=plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show(fig1)
```
