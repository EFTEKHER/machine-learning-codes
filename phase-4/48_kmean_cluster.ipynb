```python
dataset=pd.read_csv('iris-row.csv')
```

```python
dataset.head()
```

```python
sns.pairplot(data=dataset)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
```

```python
for i in range(2,21):
    km=KMeans(n_clusters=i,init='k-means++')
    km.fit(dataset)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,21)],wcss,marker='o',color='red')


plt.xticks([i for i in range(2,21)])
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid(axis='x')
plt.show()
```

```python
kmn=KMeans(n_clusters=3,init='k-means++')
dataset['predict']=kmn.fit_predict(dataset)
```

```python
dataset.head()
```

```python
dataset['predict'].value_counts()
```

```python
sns.pairplot(data=dataset,hue='predict')
```
