```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mlxtend.plotting import  plot_decision_regions
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
dataset=pd.read_csv('placement.csv')
```

```python
dataset.head()
```

```python
dataset.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x="cgpa",y="iq", data=dataset,hue="placement")
```

```python
x=dataset.iloc[:,:-1]
y=dataset['placement']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=21)
```

```python
from sklearn.svm import SVC
```

```python
sv=SVC(kernel='linear')
```

```python
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)
```

```python
sv.score(x_train,y_train)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy() , clf=sv, legend=2)
```

```python
sv1=SVC(kernel='rbf')
sv1.fit(x_train,y_train)
sv1.score(x_test,y_test)
```

```python
sv1.score(x_train,y_train)
```

```python
sv2=SVC(kernel='poly')
sv2.fit(x_train,y_train)
```

```python
sv2.score(x_test,y_test)
```

```python
sv2.score(x_train,y_train)
```

```python
sv3=SVC(kernel='sigmoid')
sv3.fit(x_train,y_train)
sv3.score(x_test,y_test)
```

```python
sv3=SVC(kernel='precomputed')
sv3.fit(x_train,y_train)
sv3.score(x_test,y_test)
```
