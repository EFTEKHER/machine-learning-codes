```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.DataFrame({"size":["s","m","l","xl","s","m","l","xl","s","m","s","m","l","xl","m","xl"]})
```

```python
df.head()
```

```python
ord_data=[["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder()
```

```python
oe=OrdinalEncoder(categories=ord_data)
```

```python
oe.fit(df[["size"]])
```

```python
oe.transform(df[["size"]])
```

```python
df['size_en']=oe.transform(df[["size"]])
```

```python
df
```

```python
ord_data1={"s":0,"m":1,"l":2,"xl":3}
```

```python
df["size_map"]=df["size"].map(ord_data1)
```
