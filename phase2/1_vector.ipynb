```python
import numpy as np 
import matplotlib.pyplot as plt;
import seaborn as sns;
```

```python
sns.set()
```

```python
# plottng a vector
```

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1 ,color='r')
plt.quiver(0,0,-3,-6, scale_units='xy', angles='xy', scale=1 ,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#addition of two vector 
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
sum= vector_1+vector_2
```

```python
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1 ,color='r')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1 ,color='y')
plt.quiver(0,0,5,1, scale_units='xy', angles='xy', scale=1 ,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
