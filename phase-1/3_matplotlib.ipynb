```python
import matplotlib.pyplot as plt

# Generate x and y values
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot of x and y')

# Display the plot
plt.show()
```

```python
import numpy as np
```

```python
x= np.linspace(0, 10, 100)
```

```python
y=np.sin(x)
z=np.cos(x)
```

```python
#plooing the data
```

```python
#sine wave

plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('sinwave')
```

```python
#cosine;
plt.plot(x,z)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('cosine wave')
plt.show()
```

```python
# adding title x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('angle');
plt.ylabel('sine value');
plt.title('sine wave');
plt.show()
```

```python
x=np.linspace(-5,5,50)
```

```python
plt.plot(x,np.sin(x),'g-');
plt.plot(x,np.cos(x),'r--');
plt.show();
```

```python
#bar plot
fig =plt.figure();
ax= fig.add_axes([0,0,1,1])
languages=['english', 'spanish', 'french', 'german', 'italian']
people=[100,200,800,50,500]
ax.bar(languages,people);
plt.xlabel("Language")
plt.ylabel("Number of People");
plt.show()
```

```python
#pie chart
fig1= plt.figure();
ax=fig1.add_axes([0,0,1,1])
languages=['english', 'spanish', 'french', 'german', 'italian']
people=[100,200,800,50,500]
ax.pie(people, labels=languages, autopct='%1.2f%%')
plt.show()
```

```python
#scatter plot;
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure();
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='r')
ax.scatter(x,z,color='g')
plt.show()
```

```python
#3D scatter plot 
fig3=plt.figure();
ax=plt.axes(projection='3d');
z=20*np.random.rand(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z ,cmap='viridis',linewidth=0.5)
plt.show()
```
