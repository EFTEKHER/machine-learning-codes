```python
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Label Encoding of breast cancer
```

```python
#loading the dataset from csv files to dataFrames
cancer_data=pd.read_csv('./breast_cancer_data.csv');
```

```python
cancer_data.head()
```

```python
#finding the count of different labels
cancer_data['diagnosis'].value_counts()
```

```python
#load the label encoder function
```

```python
label_encode=LabelEncoder()
```

```python
labels=label_encode.fit_transform(cancer_data.diagnosis)
```

```python
#appending labels to the data frame

cancer_data['target']=labels;
```

```python
cancer_data.head()
```

```python
# 0 -->Beningn # 1--->Malignant 
cancer_data['target'].value_counts()
```

```python
cancer_data['diagnosis'].value_counts()
```
