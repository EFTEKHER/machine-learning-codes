```python
#importing boston data house price data
# Download the dataset from UCI repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
column_names = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX",
    "PTRATIO", "B", "LSTAT", "MEDV"
]

# Load the dataset into a pandas DataFrame
boston_housing = pd.read_csv(url, delim_whitespace=True, names=column_names)
```

```python
import pandas as pd
import numpy as np

# Define the number of rows and columns
num_rows = 100
num_columns = 5

# Generate random data
random_data = np.random.rand(num_rows, num_columns)

# Create a DataFrame from the random data
df = pd.DataFrame(random_data, columns=[f'Column_{i+1}' for i in range(num_columns)])

# Generate random labels
labels = np.random.choice(['A', 'B', 'C'], size=num_rows)

# Add the label column to the DataFrame
df['Label'] = labels

# Define the file path
file_path = 'random_dataset_with_labels.csv'

# Save the DataFrame to a CSV file
df.to_csv(file_path, index=False)

print(f'Dataset with labels saved to {file_path}')
```
