```python
# imbalanced dataset
# A dataset with enequal class distribution 
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np;
```

```python
data=pd.read_csv('./credit_data.csv')
```

```python
data.head()
```

```python
#determine the distribution of two class
data['Class'].value_counts()
```

```python
#imblanced dataset highly 
#0-->legit 
#1-->fraud
```

```python
#separting the legit and fraudlent transactions
legit = data[data['Class']==0]
fraud = data[data['Class']==1]
```

```python
print(legit.shape)
print(fraud.shape)

```

```python
# under-sampling
# Building sample dataset containing similar distribution legit and fruad transactions
legit_sample= legit.sample(n=492)
```

```python
#concating two data frame 
new_data=pd.concat([legit_sample,fraud],axis=0)
```
