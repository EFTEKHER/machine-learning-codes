```python
import pandas as pd

# Download the dataset from UCI repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
column_names = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX",
    "PTRATIO", "B", "LSTAT", "MEDV"
]

# Load the dataset into a pandas DataFrame
boston_housing = pd.read_csv(url, delim_whitespace=True, names=column_names)

# Create a DataFrame for the Boston housing data
boston_df = pd.DataFrame(boston_housing)

# The 'MEDV' column is the target/price column
# Rename it to 'PRICE' if needed
boston_df.rename(columns={"MEDV": "PRICE"}, inplace=True)

# Display the DataFrame
print(boston_df.head())
```

```python
boston_df.head()
```

```python
#Distribution plot 
sns.displot(boston_df['PRICE'],kde=True)
```

```python
sns.distplot(boston_df['PRICE'],kde=True)
```

```python
#correlation .1positive correlation, -1 negative correlation, 0 no correlation


#constructing a heat map
```

```python
correlation = boston_df.corr()
plt.figure(figsize=(10,10))
```

```python
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
